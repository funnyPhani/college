```python
tis=[]
aus=[]
n=int(input('enter the no.of pages to scrape :'))
for i in range(1,n):
    url='http://quotes.toscrape.com/page/{}/'.format(i)
    req=requests.get(url) 
    soup=BeautifulSoup(req.text,'html.parser')
    title=soup.find_all('span',class_='text')
    for i in title:
        tis.append(i.text)
    author=soup.find_all('small',class_="author")
    
    for i in author:
        aus.append(i.text)

    
```

```python
pd.DataFrame(zip(aus,tis),columns=['Author','Title'])
```

```python
tis=[]
aus=[]
tag=[]
n=int(input('enter the no.of pages to scrape :'))
category=str(input('enter the category to scrape :'))
for i in range(1,n):
    #url='http://quotes.toscrape.com/page/{}/'.format(i)
    url='http://quotes.toscrape.com/tag/{}/page/{}/'.format(category,i)
    req=requests.get(url) 
    soup=BeautifulSoup(req.text,'html.parser')
    
    title=soup.find_all('span',class_='text')
    for i in title:
        tis.append(i.text)
    
    author=soup.find_all('small',class_="author") 
    for i in author:
        aus.append(i.text)
    
    tags=soup.find_all('div',class_='tags')
    for i in tags:
        tag.append(i.text.strip())
    
    

    
```

```python
pd.DataFrame(zip(aus,tis,tag),columns=['Author','Title',"Tags"])
```
